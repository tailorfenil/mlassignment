```python
from __future__ import division
from collections import defaultdict
import re
import math
from unidecode import unidecode
import json
import itertools
import pandas as pd

spchars = re.compile('\`|\~|\!|\@|\#|\$|\%|\^|\&|\*|\(|\)|\_|\+|\=|\\|\||\{|\[|\]|\}|\:|\;|\'|\"|\<|\,|\>|\?|\/|\.|\-')
text_file=open('F:/Documents/NYU_S3/ML/NyuPoly-CS6923-Fall16-master2/NyuPoly-CS6923-Fall16-master/Homework 3/stopwords.txt', 'r')
ListText = text_file.read().split('\n')
#print ListText
TFList=[]
IDFList=defaultdict(float)
IDF=[]
```

```python
#Function that recmove spacial character,convert charcters into lower case and remove special characters 
def make_text_parsable(text):
    # convert to unicode
    text = unidecode(text) #.decode('utf-8', 'ignore'))
    # convert text to lowercase
    text = text.lower()
    # remove special characters
    text = spchars.sub(" ", text)
    return(text)
```

```python
#function to count the number of tokens which we splits by white space without removing stopwords.We have used defaultdict(int) 
#over here which helps  us for counting words
def count_wordswithstopwords(text, wc=None):
    if wc == None:
        wc = defaultdict(int)
    tokens = text.split(" ")
    #print len(tokens)
    #tokens1=[item for item in tokens if item not in ListText]
    for t in tokens:
        wc[t] += 1
    #print wc
    return(wc)
```

```python
#function to count the number of tokens which we splits by white space with removing stopwords.We have used defaultdict(int) 
#over here which helps  us for counting words

def count_words(text, wc=None):
    if wc == None:
        wc = defaultdict(int)
    tokens = text.split(" ")
    #print len(tokens)
    tokens1=[item for item in tokens if item not in ListText]
    for t in tokens1:
        wc[t] += 1
    #print wc
    return(wc)
```

```python
#function which opens file,calls the count_words,count_wordswithstopwords,make_text_parsable function
def extract_info(filename):
    import json
    wc = defaultdict(int)
    wcwith_stop_words=defaultdict(int)
    df = defaultdict(set)
    count = 0
    bloblist=[]
    with open(filename) as fin:
        for line in fin:
            #print line
            count += 1
            current = json.loads(line)
            #print current
            text = make_text_parsable(current["abstract"] + " " + \
                current["description"] + " " + current["title"])
            #computetf(wc,len(text.split(" ")),count)
            #print text
            wcwith_stop_words=count_wordswithstopwords(text,wcwith_stop_words)
            wc = count_words(text, wc)
            #print wc
            #computetf(wc,len(text.split(" ")),count)
    #print count
    sorted_wcwithstopwords = sorted(wcwith_stop_words.items(), key=lambda x: x[1], reverse=True)
    print "(Words,Counts) with stopwords in it"
    print "\n"
    print sorted_wcwithstopwords
    sorted_wc = sorted(wc.items(), key=lambda x: x[1], reverse=True)
    #print count
    return sorted_wc
```

```python
#function to count idf
def computeidf(tw,filename):
    for word in range(0,len(tw)):
        cnt=0
        with open(filename) as fin:
            for line in fin:
                current = json.loads(line)
                text = make_text_parsable(current["abstract"] + " " + \
                current["description"] + " " + current["title"])
                if tw[word][0] in text.split():
                    cnt=cnt+1
                else:
                    continue
            IDFList[tw[word][0]]=math.log10(79/cnt)
            print ('idf for word',tw[word][0],'is ',math.log10(79/cnt))
```
